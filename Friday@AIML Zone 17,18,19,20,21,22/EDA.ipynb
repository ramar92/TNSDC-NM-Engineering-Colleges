```python
# Handling missing values:

1.detect
2. remove
3. replace

# function:
1. isnull()
2. notnull()
3. dropna()
4. fillna()
5. replace()
```

```python
df.isnull().sum()
```

```python
# replacing !!
df['Gender'].fillna("No Gender", inplace=True)

df.isnull().sum()
```

```python
df['Senior Management'].unique()
```

```python
# categarical - mode opr.
import numpy as np

mode = df['Senior Management'].mode().values[0]

df['Senior Management'] = df['Senior Management'].replace(np.nan, mode)

df.isnull().sum()
```

```python
df = df.dropna(axis=0, how='any')

print(df.isnull().sum())
df.shape
```
