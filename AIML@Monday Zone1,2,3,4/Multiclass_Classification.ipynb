```python
1. Load dataset
2. Split data
3. Select alg. SVM & DT
4. Predict the output
5. Evaluate
```

```python
from sklearn import datasets
```

```python
# load data
iris = datasets.load_iris()
```

```python
# X - input (features) | y - output

X = iris.data
y = iris.target
```

```python
# split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state= 0 )
```

What is DT?

DT - Terminologies:
1. Root Node  -Complete dataset
2. Internal node
3. Leaf Node
4. Spiliting
5. impurity *
     - Gini index
     - Entropy
6. Variance



```python
# DT :

from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier()

dt_model.fit(X_train, y_train)

dt_m_pred = dt_model.predict(X_test)
```

```python
dt_m_pred
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

pre =  tree.export_text(dt_model)
print(pre)
```

```python
fig =  plt.figure(figsize=(10,10))

_ = tree.plot_tree(dt_model, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_test, dt_m_pred)
print(cm)
```

```python
print(classification_report(y_test, dt_m_pred))
```

```python
# KNN - euclidean distance

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)
```

```python
knn_pred
```

```python
# NB :
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

gnb.fit(X_train, y_train)

gnb_pred = gnb.predict(X_test)
```

```python
gnb_pred
```

```python
print(classification_report(y_test, gnb_pred))
```

```python
#SVM

from sklearn.svm import SVC

svcmodel = SVC()
svcmodel.fit(X_train, y_train)
svc_pred = svcmodel.predict(X_test)
```

```python
svc_pred
```

```python
print(classification_report(y_test, svc_pred))
```

```python
#RF :

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train)

rf_pred = rf.predict(X_test)
```

```python
rf_pred
```

```python
print(classification_report(y_test, rf_pred))
```
